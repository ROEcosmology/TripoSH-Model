```python
values_kernels = {'000':{}, '202':{}}
for ell in values_kernels:
    for b in kernels_bk:
        if ('Pshot' in b) or ('Bshot' in b): part = 'SN'
        else: part = 'tree'
        values_kernels[ell][b] = np.load(f'/home/rneveux/kernels/bk/z.8/k_1_20/no_f/{ell}/{part}/{b}.npy', allow_pickle=True).item()['K']
```

```python
bias_params = {'b1': 1.9943274e+00,
 'b2': -2.8493848e-01,
 'bG2': -2.9738286e-01,
 'bGamma3': 6.8452853e-01,
 'c0': 1.1104226e+01,
 'c2pp': 3.0399105e+01,
 'c4pp': 3.0399105e+01,
 'c1': 5.9360800e+00,
 'c2': -1.6427413e-01,
 'ch': 3.9950281e+02,
 'Pshot': 5.6773871e-01,
 'a0': -4.3189287e+00,
 'Bshot': 2.9556961e+00}

f_fid = .8382695826838665
```

```python
def rebuild_bk(bias_params, ell, mean_density=1e-3):
    
    bk_model = np.zeros_like(values_kernels[ell]['b1_b1_b1'])
    for b, values in values_kernels[ell].items():
        bias = 1

        if 'b1' in b:
            bias *= bias_params['b1'] ** b.count('b1')
        if 'b2' in b:
            bias *= bias_params['b2']
        if 'bG2' in b:
            bias *= bias_params['bG2']
        if 'c1' in b:
            bias *= bias_params['c1'] ** b.count('c1')
        if 'c2' in b:
            bias *= bias_params['c2'] ** b.count('c2')
        if 'Pshot' in b:
            bias *= (1 + bias_params['Pshot']) / mean_density
        if 'Bshot' in b:
            bias *= bias_params['Bshot'] / mean_density
        if 'f' in b:
            bias *= bias_params['f'] ** b.count('f')

        bk_model += bias * values

    if bias_params['Pshot']!=0: bk_model += ((1+bias_params['Pshot'])/mean_density)**2
    
    return bk_model
```

```python
for f in np.linspace(.6,1,10):
    bias_params['f'] = f
    for ell in ['000','202']:
        bk = rebuild_bk(bias_params, ell, mean_density=1e-3)
        c = (f-.6)/.4
        if ell == '000' and (f==.6 or f==1.): label = r'f='+str(f)
        else: label=None
        plt.plot(k, k**2*bk, c=[1-c,1-c,c], alpha=.5, label=label)
bias_params['f'] = f_fid
for ell in ['000','202']:
    bk = rebuild_bk(bias_params, ell, mean_density=1e-3)
    if ell == '000': label = r'f_{fid}='+"{:.2f}".format(f_fid)
    else: label=None
    plt.plot(k, k**2*bk, c='k', label=label)
plt.xlabel(r'$k[h\mathrm{Mpc}^{-1}]$', fontsize=16)
plt.ylabel(r'$k^2\cdot B(k)[h^{-1}\mathrm{Mpc}]^4$', fontsize=16)
plt.legend(fontsize=16)
```
