```python
kernels = {}
kernels_sn = {}
kernels_png = {}
kernels_ct = {}
for i in group[5]:
    kernels[i] = np.load(f'/home/rneveux/kernels/bk/z.8/k_1_20/202/tree/{i}.npy', allow_pickle=True).item()
for i in group[0]:
    kernels_ct[i] = np.load(f'/home/rneveux/kernels/bk/z.8/k_1_20/202/tree/{i}.npy', allow_pickle=True).item()
for i in sn:
    kernels_sn[i] = np.load(f'/home/rneveux/kernels/bk/z.8/k_1_20/202/SN/{i}.npy', allow_pickle=True).item()
#for ell in ['000','202']:
kernels_png = {}
for i in png:
    kernels_png[i] = np.load(f'/home/rneveux/kernels/bk/z.8/k_1_20/202/PNG/{i}.npy', allow_pickle=True).item()
```

```python
for i in ['b1_b1_f']:#'b1_b1_b2']:
    plt.plot(kernels[i]['kbin'], kernels[i]['kbin']**2*kernels[i]['K'], label='tree') #, color='k')
for i in ['c1_b1_b2']:
    if kernels_ct[i]['K'][0]<0:
        kernels_ct[i]['K']*= -1
        linestyle = ':'
    else:
        linestyle = '-'
    plt.plot(kernels_ct[i]['kbin'], kernels_ct[i]['kbin']**2*kernels_ct[i]['K'], label='counterterms', color='g', linestyle=linestyle)
for i in ['Pshot_f_b1']: #['Bshot_b1_b1']:
    plt.plot(kernels_sn[i]['kbin'], kernels_sn[i]['kbin']**2*kernels_sn[i]['K']*1e3, label='shotnoise', color='r')
for i in ['fnlequi_b1_b1_f', 'fnlortho_LSS_b1_b1_f']: #['fnlequi_b1_b1_b1', 'fnlortho_LSS_b1_b1_b1']:
    if kernels_png[i]['K'][0]<0:
        kernels_png[i]['K']*= -1
        if 'equi' in i:linestyle = ':'
        if 'ortho' in i:linestyle = ':'
    else:
        linestyle = '-'
    if 'ortho' in i: color = 'cyan'
    else: color = 'b'
    plt.plot(kernels_png[i]['kbin'], kernels_png[i]['kbin']**2*kernels_png[i]['K'], label=i[3:7], color=color, linestyle=linestyle)
plt.legend()
plt.xlabel('$k$', fontsize=20)
plt.ylabel(r'$k^2B(k)$', fontsize=20)
plt.yscale('log')
```
