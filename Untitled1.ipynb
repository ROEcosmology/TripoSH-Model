```python
import os
from classy import Class
import numpy as np
from powerbispectrum import ComputePowerBiSpectrum
from matplotlib import pyplot as plt
import matplotlib.cm as cm
plt.rcParams['figure.figsize'] = [15, 10]
```

```python
h_fid = .6736
omega_b_fid = .02237
omega_cdm_fid = .12
n_s_fid = .9649
A_s_fid = 2.083e-9

params_cosmo = {
            'output': 'mPk',
            'h': h_fid,
            'omega_b': omega_b_fid,
            'omega_cdm': omega_cdm_fid,
            'n_s': n_s_fid,
            'A_s': A_s_fid,
            'tau_reio': 0.0544,
            'N_ncdm': 1.,
            'm_ncdm': 0.06,
            'N_ur': 2.0328,
            'z_max_pk': 3.,
            'P_k_max_h/Mpc': 50.,
            }
z = .8
```

```python
k = np.arange(.01, .2, .002)
len(k)
```

```python
max_a_post = {'b1': 1.9599205896482994,
 'b2': 2.153378104444724,
 'bG2': 1.8531626151434741,
 'c1': 3.5839809245200556,
 'c2': 6.344556262145687,
 'ch': 520.3124187220282,
 'Pshot': 2,
 'Bshot': 2}
```

```python
cl = ComputePowerBiSpectrum(params_cosmo, z, diag=True)
cl.initial_power_spectrum()
```

```python
%%time
ell1 = 2
ell2 = 0
ELL = 2
cl.calc_B_diag(
        k, ell1, ell2, ELL,
        alpha_perp=1, alpha_parallel=1, b1=max_a_post['b1'], b2=max_a_post['b2'], bG2=max_a_post['bG2'],
        c1=max_a_post['c1'], c2=max_a_post['c2'], knl=.3,
        Pshot=max_a_post['Pshot']*1000, Bshot=max_a_post['Bshot']*1000,
        integrand='tree',
        ks=.05,
    )
tree = cl.BK
cl.calc_B_diag(
        k, ell1, ell2, ELL,
        alpha_perp=1, alpha_parallel=1, b1=max_a_post['b1'], b2=max_a_post['b2'], bG2=max_a_post['bG2'],
        c1=max_a_post['c1'], c2=max_a_post['c2'], knl=.3,
        Pshot=max_a_post['Pshot']*1000, Bshot=max_a_post['Bshot']*1000,
        integrand='SN',
        ks=.05,
    )
SN = cl.BK
BK = tree['K']+SN['K']
```

```python
%%time
ell1 = 2
ell2 = 0
ELL = 2
cl.calc_B_diag(
        k, ell1, ell2, ELL,
        alpha_perp=1, alpha_parallel=1, b1=max_a_post['b1'], b2=max_a_post['b2'], bG2=max_a_post['bG2'],
        c1=max_a_post['c1'], c2=max_a_post['c2'], knl=.3,
        Pshot=max_a_post['Pshot']*1000, Bshot=max_a_post['Bshot']*1000,
        integrand='tree',
        ks=.05,
        f = 1
    )
tree_ = cl.BK
cl.calc_B_diag(
        k, ell1, ell2, ELL,
        alpha_perp=1, alpha_parallel=1, b1=max_a_post['b1'], b2=max_a_post['b2'], bG2=max_a_post['bG2'],
        c1=max_a_post['c1'], c2=max_a_post['c2'], knl=.3,
        Pshot=max_a_post['Pshot']*1000, Bshot=max_a_post['Bshot']*1000,
        integrand='SN',
        ks=.05,
        f = 1
    )
SN_ = cl.BK
BK_ = tree_['K']+SN_['K']
```

```python
plt.plot(tree['kbin'], (BK-BK_)/BK_)
plt.plot(tree['kbin'], (tree['K']-tree_['K'])/tree_['K'])
plt.plot(tree['kbin'], (SN['K']-SN_['K'])/SN_['K'])
```

```python
plt.plot(tree['kbin'], tree['kbin']**2*BK)
plt.plot(tree['kbin'], tree['kbin']**2*BK_)

plt.plot(tree['kbin'], tree['kbin']**2*tree['K'])
plt.plot(tree['kbin'], tree['kbin']**2*tree_['K'])

plt.plot(tree['kbin'], tree['kbin']**2*SN['K'])
plt.plot(tree['kbin'], tree['kbin']**2*SN_['K'])
```

```python
for i in [0,2]:
    plt.plot(tree['kbin'], tree['kbin']**i*BK, label='tot')
    plt.plot(tree['kbin'], tree['kbin']**i*tree['K'], label='tree')
    plt.plot(tree['kbin'], tree['kbin']**i*SN['K'], label='SN')
    plt.plot(tree['kbin'], tree['kbin']**i*BK_, label='tot now')
    plt.plot(tree['kbin'], tree['kbin']**i*tree_['K'], label='tree now')
    plt.plot(tree['kbin'], tree['kbin']**i*SN_['K'], label='SN now')
    plt.legend()
    plt.show()
```
